```python
#removing skewness using power transform
from sklearn.preprocessing import PowerTransformer
```

```python
pt=PowerTransformer(method='yeo-johnson')
```

```python
data_new=pt.fit_transform(data)
```

```python
happiness=pd.DataFrame(data=data_new,columns=data.columns)
```

```python
#outliers are removed from the columns after removing the skewness.
```

```python
x=happiness.drop('Happiness Rank',axis=1)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
x=pd.DataFrame(data=x,columns=('Country','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom'))
```

```python
x
```

```python
y=data['Happiness Rank']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.metrics import r2_score
list=[LinearRegression(),SVR(kernel='linear'),DecisionTreeRegressor(),KNeighborsRegressor(),RandomForestRegressor(),AdaBoostRegressor()]
for j in list:
    r_max=0
    print('for the algorithm: ',j)
    for i in range(33,101):
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
        j.fit(x_train,y_train)
        pred=j.predict(x_test)
        r2_scr=r2_score(pred,y_test)
        print('r2 score corresponding to random state',i,'is: ',r2_scr)
        if r2_scr>r_max:
            r_max=r2_scr
            final=i
    print('\n')
    print('Max r2 score corresponding to: ', final,'is: ',r_max)
    print('\n')
```

```python
param = {
    'kernel':['linear','poly','rbf']
}



svr=SVR()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
cv = GridSearchCV(svr, param_grid=param)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=70)
cv.fit(x_train, y_train)
print(cv.best_score_)
print(cv.best_estimator_)
print(cv.best_params_)
```

we observed in grid search cv that the model is performing better with kernel=linear. So we proceed further and save this model into joblib.

```python
from sklearn.metrics import mean_squared_error
import numpy as np
sv=SVR(kernel='linear')
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=70)
sv.fit(x_train,y_train)
pred=sv.predict(x_test)
print('R2 score: ', r2_score(pred,y_test))
print('Mean Squared Error: ',mean_squared_error(pred,y_test))
print('Root Mean Squared Error: ', np.sqrt(mean_squared_error(pred,y_test)))
```
